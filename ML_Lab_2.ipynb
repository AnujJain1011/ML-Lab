```python
import matplotlib.pyplot as plt
x=[2,8,9,10]
y=[11,16,23,34]

x2=[3,9,11,14]
y2=[6,15,7,17]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='red')
plt.title('Bar Graph')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```
